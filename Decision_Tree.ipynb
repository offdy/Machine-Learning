Decision tree
------
* 학습 데이터로부터 결정 나무를 학습
* 비교적 해석 가능한 형태의 모델을 얻을 수 있음
* max depth등의 hyperparameter 설정을 필요로 함
* Overfitting이 쉽게 발생
* Regression을 위해서도 사용 가능
* 정확도를 높이기 위해 여러 decision tree를 같이 사용하는 random forest가 자주 사용됨
* Overfitting을 해결하기 위해 Pruning(가지치기) 기술이 사용


```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.tree import DecisionTreeClassifier
from sklearn import datasets
```

```python
iris = datasets.load_iris()
X = iris.data
y = iris.target
```

```python
model = DecisionTreeClassifier(max_depth=4)
model.fit(X, y)
```

```python
from sklearn import tree
plt.figure(figsize=(15,15))
tree.plot_tree(model)
plt.show()
```
